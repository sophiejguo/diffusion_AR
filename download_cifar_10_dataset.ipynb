```python
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import CIFAR10
import os
import shutil
from tqdm import tqdm

def save_cifar10_as_imagenet_format(root_dir="cifar10_imagenet_format", train_split=0.9):
    # Download dataset
    transform = transforms.ToTensor()
    train_set = CIFAR10(root="./data", train=True, download=True, transform=transform)
    test_set = CIFAR10(root="./data", train=False, download=True, transform=transform)

    # Combine and shuffle
    data = [(img, label, train_set.classes[label]) for img, label in train_set] + \
           [(img, label, train_set.classes[label]) for img, label in test_set]

    # Shuffle for split
    import random
    random.seed(42)
    random.shuffle(data)

    # Split
    split_idx = int(len(data) * train_split)
    train_data, val_data = data[:split_idx], data[split_idx:]

    def save_images(split_name, split_data):
        for idx, (img, label, class_name) in enumerate(tqdm(split_data, desc=f"Saving {split_name}")):
            class_dir = os.path.join(root_dir, split_name, class_name)
            os.makedirs(class_dir, exist_ok=True)
            img_path = os.path.join(class_dir, f"{idx}.png")
            torchvision.utils.save_image(img, img_path)

    save_images("train", train_data)
    save_images("val", val_data)

if __name__ == "__main__":
    save_cifar10_as_imagenet_format()
```
